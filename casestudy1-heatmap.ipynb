This first dataset is about the start station for casual riders during the weekends for the period of analysis (04/2021-03/2022).
The goal is to verify the area with most potential for advertisement (where most casual riders start the bike ride during the weekend).

```python
# read in the data 
casual_start_station_weekend = pd.read_csv("/kaggle/input/start-station-casual-weekend/start_station_casual_weekend.csv")
```

```python
#checking the data
casual_start_station_weekend.head()
#type(casual_start_station_weekend)
```

## Creating the GeoDataFrame
Now we are going to use the latitude and longitude to create a GeoDataFrame

```python
# Convert the DataFrame to a GeoDataFrame
gpd_start_station_casual_weekend = gpd.GeoDataFrame(
    casual_start_station_weekend,
    geometry=gpd.points_from_xy(casual_start_station_weekend.start_lng, casual_start_station_weekend.start_lat))

# Set the coordinate reference system (CRS) to EPSG 4326
gpd_start_station_casual_weekend.crs = {'init': 'epsg:4326'}

gpd_start_station_casual_weekend.head()
```

```python
print(gpd_start_station_casual_weekend.crs)
```

Importing the necessary packages to create the heatmap

```python
import folium
from folium import Choropleth, Circle, Marker
from folium.plugins import HeatMap, MarkerCluster
```

## Creating the base map of Chicago

```python
# Create a base map
Chicago_map = folium.Map(location=[41.87,-87.63], tiles='cartodbpositron', zoom_start=13)

Chicago_map
```

## Adding the heatmap
The heatmap represent the locations of the start stations the casual users collect the bike on weekends

```python
# Add a heatmap to the base map
HeatMap(data=gpd_start_station_casual_weekend[['start_lat', 'start_lng']], radius=11).add_to(Chicago_map)

Chicago_map
```

The above interactive plot shows that there is a bigger concentration of casual riders starting theirs rides during the weekends in the areas of navy pier, lakefront trail, lake shore and millenium park. These areas are good for a marketing campaign aiming casual riders.
